```python
from __future__ import division
from sympy import *
x = symbols('x')
C1, C2, C3, C4 = symbols('C1, C2, C3, C4')
theta1, theta2, phi1, phi2 = symbols('theta1, theta2, phi1, phi2')
k, mt1, mt2, L, EIw, GJ = symbols("k, mt1, mt2, L, EIw, GJ")
theta = symbols("theta", cls=Function)
```

```python
theta = C1 + C2 * x + C3 * cosh(k * x) + C4 * sinh(k * x)
```

```python
phi = diff(theta, x)
Bw = -EIw * diff(theta, x, x)
Ts = GJ * diff(theta, x)
Tw = diff(Bw, x)
T = Ts + Tw
```

```python
eq_1 = Eq(lhs=theta, rhs=theta1).subs({x:0})
eq_2 = Eq(lhs=phi, rhs=phi1).subs({x:0})
eq_3 = Eq(lhs=theta, rhs=theta2).subs({x:L})
eq_4 = Eq(lhs=phi, rhs=phi2).subs({x:L})
```

```python
val = solve((eq_1, eq_2, eq_3, eq_4), ('C1, C2, C3, C4'))
```

```python
theta = theta.subs(val)
phi = phi.subs(val)
Ts = Ts.subs(val)
Tw = Tw.subs(val)
T = T.subs(val)
Bw = Bw.subs(val)
```
